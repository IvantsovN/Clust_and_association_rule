# Ассоциативные правила 
Скрипт позволяет находить в массиве данных ассоциативные правила. 

```python
# импортируем библиотеки 
import pandas as pd
import numpy as np
```

```python
#загружаем данные их файла MS Excel
sales=pd.read_excel('DataSet.xlsx')
#проверяем результат, выводим первые 10 строчек кода
sales.head(10)
```

```python
#преобразуем таблицу таким образом, чтобы столбцы были наименования ассортимента, строки клиенты, а на пересечении
# рейтинг
UID_sales=sales.pivot_table(values='rating', columns=['genres'], index='user_id', aggfunc=np.sum)
#Вместо NaN (значение отсутствует) проставим в ДатаФрейм 0-ки:
UID_sales[np.isnan(UID_sales)] = 0
# все значения больше 0 заменим 1, чтобы больше проявить правила.
UID_sales[UID_sales>0]=1
# 10 первых строк новой таблицы
UID_sales.head(10)
```

```python
#Создадим функцию для создания списка транзакций:
def transaction_list(df):
    list_external=[]
    for i in range(df.shape[0]):
        list_internal=[]
        data=df.iloc[i]
        index=data[data>0]
        for element in index.index:
            list_internal.append(element)
        list_external.append(list_internal)
    return list_external
# используем функцию, чтобы преобразовать массив в список транзакций

transactions=transaction_list(UID_sales[:5000])
# в данном случае не обращаем внимание на предупреждение
# как выглядит наш список
# Безусловно, к нему можно было придти и не строя бинарную матрицу
# но с бинарной матрицей можно выполнять другие операции, например, искать теже кластера
# это список товаров, которые содержаться в чеке или их покупает клиент
transactions[0]
```

Для использования библиотеки mlxtend ее следует установить, она не является стандартной. Установка через Anaconda Prompt командой pip install --upgrade mlxtend

```python
# загрузим пакеты, необходимые для выполнения анализа
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

Создадим частные наборы с уровнем поддержки (min_support) равной 0.2 Это крайне низкий уровень поддержки для практического применения, используется исключительно для примера.

```python
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
df = pd.DataFrame(te_ary, columns=te.columns_)
frequent_itemsets = apriori(df, min_support=0.2, use_colnames=True)
```

```python
frequent_itemsets
```

```python
# сгенерируем ассоциативные правила с уровнем доверия 0.1
association_rules(frequent_itemsets, metric="confidence", min_threshold=0.7)
```

```python
# найдем ассоциативные правила с уровнем независимости больше 1.7
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.7)
rules
```

```python
#имя файла AssocRules002.xlsx
writer = pd.ExcelWriter('association_rule.xlsx')
rules.to_excel(writer,'AR')
writer.save()
```
